* randn(-int to + int) : value between positive number to positive number or vice versa
* randint( ) : integer vales between given range

## array arithmatics

```python
arr1 = np.array([1,2,3,4])
arr2 = np.array([3,4,2,1])
print(f"addition : {arr1+arr2}")
print(f"subtraction : {arr1-arr2}")
print(f"multiplication : {arr1*arr2}")
print(f"division : {arr1//arr2}")
```

```python
arr2D_1 = np.array([[1,2,3,4],[1,5,3,2]])
arr2D_2 = np.array([[1,2,3,8],[1,5,3,2]])
print(f"addition : {arr2D_1+arr2D_2}")
print(f"subtraction : {arr2D_1-arr2D_2}")
print(f"multiplication : {arr2D_1*arr2D_2}")
print(f"division : {arr2D_1//arr2D_2}")
print(f"remainder : {arr2D_1%arr2D_2}")


```

## statistical analysis

```python
# mean
print(np.mean(arr1))
print(np.sum(arr1))
print(np.std(arr1))
print(np.var(arr1))
print(np.min(arr1))
print(np.max(arr1))
print(np.percentile(arr1,75))
print(np.sort(arr1))
print(np.unique(arr1))
print(np.sqrt(arr1))




```

```python
# search and filter
print(arr1)
print(np.where(arr1>2))
```

```python
# copy and views
arr1_copy = arr1.copy()

arr1_copy[1]= 10
print(arr1_copy)
print(arr1)
```

```python
arr1_view = arr1.view()
arr1_view[1] = 14
print(arr1_view)
print(arr1)
```
