```python
import numpy as np
```

```python
array_1d = np.array([1,2,3,4])
```

## 1D array

```python
print(array_1d)
```

## 2D array

```python
array_2d = np.array([[1,2],[2,9]])
print(array_2d)
```

## 3D array


```python
array_3d = np.array([[[1,2,3],[2,3,4]],
                     [[3,4,5],[8,9,0]]])
print(array_3d)
```

## Check `datatype` of array

```python
print(array_1d.dtype)
```

## check `dimension` of array

```python
print(array_3d.ndim)
```

## Check `shape` of array

```python
print(array_3d.shape)
```

## check `item size` of array

```python
print(array_3d.itemsize)
```

## create array with zeroes

```python
print(np.zeros((2,3)).dtype)
```

```python
print(np.zeros((2,3,3)))
```

```python
print(np.ones((2,3)))
```
